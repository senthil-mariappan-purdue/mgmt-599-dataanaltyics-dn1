```python
#@title Import Python Libraries & Some Other Setup
# Basic Python data science libraries
import pandas as pd
import numpy as np
import scipy.optimize

# Import and setup for plotly in Colab
import chart_studio
import chart_studio.plotly as py
import plotly.graph_objects as go
import plotly.io as pio
import plotly.express as px

# Enable displaying pandas data frames as interactive tables by default
from google.colab import data_table
data_table.enable_dataframe_formatter()
```

```python

querynew = """
    SELECT *
    FROM `mgmt599-ajaygopalakrishnan.Superstorelab1AJ.superstoreAJAY_table`
"""
client = bigquery.Client(project='mgmt599-ajaygopalakrishnan')
# Run the query and convert to a pandas DataFrame
df = client.query(querynew).to_dataframe()

# Display the results
df.tail()
```

```python
display(df.describe(include='all'))
```

```python
df.info()
```

```python
df['zip'] = df['zip'].astype(object)
df.info()
```

```python
df.isnull().sum()
```

```python
zip_city_overlap = df.groupby('zip')['city'].nunique().reset_index(name='unique_cities')
overlapping_zips = zip_city_overlap[zip_city_overlap['unique_cities'] > 1]

if overlapping_zips.empty:
    print("No cities overlap on zip code.")
else:
    print("The following zip codes are associated with more than one city:")
    display(overlapping_zips)
```

```python
category_sales_profit = df.groupby('category').agg(
    total_sales=('sales', 'sum'),
    total_profit=('profit', 'sum')
)

category_sales_profit['profit_margin'] = category_sales_profit['total_profit'] / category_sales_profit['total_sales']

top_categories_by_sales = category_sales_profit.sort_values(by='total_sales', ascending=False)

display(top_categories_by_sales)
```

```python
fig = px.bar(top_categories_by_sales, y='total_sales', title='Total Sales by Category')
fig.show()
```

```python
region_sales = df.groupby('region')['sales'].sum().reset_index()
region_sales = region_sales.sort_values(by='sales', ascending=False)

fig = px.bar(region_sales, x='region', y='sales', title='Total Sales by Region')
fig.show()
```

```python
region_state_city_sales = df.groupby(['region', 'state', 'city'])['sales'].sum().reset_index()

fig = px.treemap(region_state_city_sales, path=['region', 'state', 'city'], values='sales',
                 title='Sales by Region, State, and City')
fig.show()
```

```python
region_analysis = df.groupby('region').agg(
    total_orders=('order_id', 'nunique'),
    total_quantity=('quantity', 'sum'),
    total_profit=('profit', 'sum'),
    average_profit_margin=('profit_margin', 'mean')
).reset_index()

region_analysis['profit_per_order'] = region_analysis['total_profit'] / region_analysis['total_orders']
region_analysis['profit_per_quantity'] = region_analysis['total_profit'] / region_analysis['total_quantity']

display(region_analysis)
```

```python
import plotly.express as px

# Scatter plot of Quantity vs. Sales
fig_quantity_sales = px.scatter(df, x='quantity', y='sales', title='Quantity vs. Sales')
fig_quantity_sales.show()

# Scatter plot of Quantity vs. Profit
fig_quantity_profit = px.scatter(df, x='quantity', y='profit', title='Quantity vs. Profit')
fig_quantity_profit.show()
```

```python
import plotly.express as px

# Scatter plot of Discount vs. Sales
fig_discount_sales = px.scatter(df, x='discount', y='sales', title='Discount vs. Sales')
fig_discount_sales.show()

# Scatter plot of Discount vs. Profit
fig_discount_profit = px.scatter(df, x='discount', y='profit', title='Discount vs. Profit')
fig_discount_profit.show()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])

daily_sales_profit = df.groupby('order_date').agg(
    total_sales=('sales', 'sum'),
    total_profit=('profit', 'sum')
).reset_index()

daily_sales_profit = daily_sales_profit.sort_values(by='order_date')

fig_sales_time = px.line(daily_sales_profit, x='order_date', y='total_sales', title='Total Sales Over Time')
fig_sales_time.show()

fig_profit_time = px.line(daily_sales_profit, x='order_date', y='total_profit', title='Total Profit Over Time')
fig_profit_time.show()
```

```python
customer_sales_profit = df.groupby('customer').agg(
    total_sales=('sales', 'sum'),
    total_profit=('profit', 'sum')
).reset_index()

customer_sales_profit_sorted_sales = customer_sales_profit.sort_values(by='total_sales', ascending=False)
customer_sales_profit_sorted_profit = customer_sales_profit.sort_values(by='total_profit', ascending=False)


print("Top 10 Customers by Sales:")
display(customer_sales_profit_sorted_sales.head(10))

print("\nTop 10 Customers by Profit:")
display(customer_sales_profit_sorted_profit.head(10))
```

```python
order_sales_profit = df.groupby('order_id').agg(
    total_sales=('sales', 'sum'),
    total_profit=('profit', 'sum')
).reset_index()

print("Descriptive statistics for Sales per Order:")
display(order_sales_profit['total_sales'].describe())

print("\nDescriptive statistics for Profit per Order:")
display(order_sales_profit['total_profit'].describe())
```
